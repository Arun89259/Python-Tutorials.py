```python
# Q3. Print odd numbers from 1 to 20

for i in range(1, 21, 2):
    print(i)
```

```python
# Q4. Print numbers from 10 to 1

for i in range(10, 0, -1):
    print(i)
```

```python
# Q5. Multiplication table of 5

num = 5
for i in range(1, 11):
    print(num, "x", i, "=", num*i)
```

```python
# Q6. Sum of numbers from 1 to 100

total = 0
for i in range(1, 101):
    total += i

print("Sum =", total)
```

```python
# Q7. Sum of even numbers from 1 to 50

total = 0
for i in range(2, 51, 2):
    total += i

print("Sum =", total)
```

```python
# Q8. Print list elements

my_list = [10, 20, 30, 40, 50]

for item in my_list:
    print(item)
```

```python
# Q9. Count elements without len()

my_list = [10, 20, 30, 40, 50]

count = 0
for item in my_list:
    count += 1

print("Count =", count)
```

```python
# Q10. Largest number in list

my_list = [10, 50, 20, 90, 30]

largest = my_list[0]

for num in my_list:
    if num > largest:
        largest = num

print("Largest =", largest)
```

```python
# Q11. Count numbers divisible by 3 (1–100)

count = 0

for i in range(1, 101):
    if i % 3 == 0:
        count += 1

print("Total =", count)
```

```python
# Q12. Numbers divisible by both 3 and 5

for i in range(1, 51):
    if i % 3 == 0 and i % 5 == 0:
        print(i)
```

```python
# Q13. Factorial

num = 5
fact = 1

for i in range(1, num+1):
    fact *= i

print("Factorial =", fact)
```

```python
# Q14. Reverse a number

num = 1234
rev = 0

for digit in str(num):
    rev = rev*10 + int(digit)

print("Reverse =", rev)
```

```python
# Q15. Print each character of string

text = "hello"

for ch in text:
    print(ch)
```

```python
# Q16. Count vowels

text = "education"
count = 0

for ch in text.lower():
    if ch in "aeiou":
        count += 1

print("Vowels =", count)
```

```python
# Q17. Count digits in number

num = 12345
count = 0

for digit in str(num):
    count += 1

print("Digits =", count)
```

```python
# Q20. Sum of digits

num = 123
total = 0

for digit in str(num):
    total += int(digit)

print("Sum =", total)
```

```python
# Q21. Star pattern

for i in range(1, 6):
    print("*"*i)
```

```python
# Q22. Number pattern

for i in range(1, 5):
    for j in range(1, i+1):
        print(j, end="")
    print()
```

```python
# Q23. Reverse star pattern

for i in range(5, 0, -1):
    print("*"*i)
```
