```
# Q1: Print numbers from 1 to 10 using a for loop
for i in range(1, 11):
    print(i)
```

```
# Q2: Print even numbers from 1 to 20
for i in range(1, 21):
    if i % 2 == 0:
        print(i)
```

```
# Q3: Print odd numbers from 1 to 20
for i in range(1, 21):
    if i % 2 != 0:
        print(i)
```

```
# Q4: Print numbers from 10 to 1
for i in range(10, 0, -1):
    print(i)
```

```
# Q5: Multiplication table of 5
n = 5
for i in range(1, 11):
    print(n, 'x', i, '=', n*i)
```

```
# Q6: Sum from 1 to 100
total = 0
for i in range(1, 101):
    total += i
print(total)
```

```
# Q7: Sum of even numbers from 1 to 50
total = 0
for i in range(1, 51):
    if i % 2 == 0:
        total += i
print(total)
```

```
# Q8: Print elements of a list
lst = [10, 20, 30, 40, 50]
for i in lst:
    print(i)
```

```
# Q9: Count elements without len()
lst = [10, 20, 30, 40, 50]
count = 0
for i in lst:
    count += 1
print(count)
```

```
# Q10: Find largest number in list
lst = [10, 20, 5, 40, 30]
largest = lst[0]
for i in lst:
    if i > largest:
        largest = i
print(largest)
```

```
# Q11: Count numbers divisible by 3 (1-100)
count = 0
for i in range(1, 101):
    if i % 3 == 0:
        count += 1
print(count)
```

```
# Q12: Numbers divisible by 3 and 5 (1-50)
for i in range(1, 51):
    if i % 3 == 0 and i % 5 == 0:
        print(i)
```

```
# Q13: Factorial using for loop
n = 5
fact = 1
for i in range(1, n+1):
    fact *= i
print(fact)
```

```
# Q14: Reverse number
num = 1234
rev = 0
while num > 0:
    rev = rev * 10 + num % 10
    num //= 10
print(rev)
```

```
# Q15: Print each character
s = "hello"
for ch in s:
    print(ch)
```

```
# Q16: Count vowels
s = "hello world"
count = 0
for ch in s:
    if ch in 'aeiouAEIOU':
        count += 1
print(count)
```

```
# Q17: Count digits
num = 12345
count = 0
while num > 0:
    count += 1
    num //= 10
print(count)
```

```
# Q18: Square of 1 to 10
for i in range(1, 11):
    print(i*i)
```

```
# Q19: Cube of 1 to 10
for i in range(1, 11):
    print(i**3)
```

```
# Q20: Sum of digits
num = 123
sum_d = 0
while num > 0:
    sum_d += num % 10
    num //= 10
print(sum_d)
```
