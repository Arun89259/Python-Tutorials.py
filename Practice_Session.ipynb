# BASIC LEVEL

```python
# Q1   Print numbers from 1 to 50. For each number:
#    * If divisible by 3 → print "Fizz"
#    * If divisible by 5 → print "Buzz"
#    * If divisible by both → print "FizzBuzz"


for i in range(1, 51):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

```python
# 2. Take a number n and print whether each number from 1 to n is even or odd.

n = 10

for i in range(1, n+1):
    if i % 2 == 0:
        print(i, "Even")
    else:
        print(i, "Odd")
```

```python
# 3. Print all prime numbers between 1 to 100.

for num in range(2, 101):
    prime = True
    for i in range(2, num):
        if num % i == 0:
            prime = False
            break
    if prime:
        print(num)
```

```python
# 4. Take a number and count how many digits are even and how many are odd.

num = 123456
even = 0
odd = 0

for d in str(num):
    if int(d) % 2 == 0:
        even += 1
    else:
        odd += 1

print("Even digits:", even)
print("Odd digits:", odd)
```

```python
# 5. Print multiplication table of a number using for loop
#    and if condition to highlight multiples of 5.

num = 7

for i in range(1, 11):
    if i % 5 == 0:
        print(num*i, "<-- multiple of 5")
    else:
        print(num*i)
```

```python
# 6. Reverse a number using a while loop.

num = 1234
rev = 0

while num > 0:
    digit = num % 10
    rev = rev*10 + digit
    num //= 10

print(rev)
```

```python
# 7. Check whether a number is palindrome using loop and condition.

num = 121
temp = num
rev = 0

while num > 0:
    rev = rev*10 + num % 10
    num //= 10

if temp == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# 8. Find the sum of all numbers from 1 to n,
#    but skip numbers divisible by 7.

n = 50
total = 0

for i in range(1, n+1):
    if i % 7 == 0:
        continue
    total += i

print(total)
```

```python
# 10. Find the factorial of a number using while loop.

n = 5
fact = 1
i = 1

while i <= n:
    fact *= i
    i += 1

print(fact)
```

```python
# 11. Count frequency of vowels and consonants in a string
#     using loop and condition.

text = "education"
vowel = 0
cons = 0

for ch in text.lower():
    if ch.isalpha():
        if ch in "aeiou":
            vowel += 1
        else:
            cons += 1

print("Vowels:", vowel)
print("Consonants:", cons)
```

```python
# 12. Take a list of numbers:
#     Print only positive even numbers
#     Skip negative numbers using continue

nums = [10,-2,15,20,-5,8]

for n in nums:
    if n < 0:
        continue
    if n % 2 == 0:
        print(n)
```

```python
# 13. Find the largest and smallest number in a list using loop.

nums = [10,50,2,99,4]

largest = nums[0]
smallest = nums[0]

for n in nums:
    if n > largest:
        largest = n
    if n < smallest:
        smallest = n

print("Largest:", largest)
print("Smallest:", smallest)
```

```python
# 14. Print all Armstrong numbers between 1 to 1000.

for num in range(1,1000):
    total = 0
    power = len(str(num))

    for d in str(num):
        total += int(d)**power

    if total == num:
        print(num)
```

```python
# 15. Take a number and check if it is perfect number.

num = 28
total = 0

for i in range(1,num):
    if num % i == 0:
        total += i

if total == num:
    print("Perfect Number")
else:
    print("Not Perfect")
```

```python
# 16. Create a menu-driven program using while loop:
#     1 → Add two numbers
#     2 → Subtract
#     3 → Exit

while True:

    print("1 Add")
    print("2 Subtract")
    print("3 Exit")

    choice = int(input("Choice: "))

    if choice == 3:
        break

    a = int(input("A: "))
    b = int(input("B: "))

    if choice == 1:
        print(a+b)
    elif choice == 2:
        print(a-b)
```

```python
# 18. Take a list and:
#     Count even, odd, positive, negative numbers using one loop

nums = [10,-5,3,-2,8,7]

even = odd = pos = neg = 0

for n in nums:
    if n % 2 == 0:
        even += 1
    else:
        odd += 1

    if n > 0:
        pos += 1
    else:
        neg += 1

print("Even:", even)
print("Odd:", odd)
print("Positive:", pos)
print("Negative:", neg)
```

```python
# 19. Print a pyramid pattern:
#        *
#       ***
#      *****

for i in range(1,4):
    print(" "*(3-i) + "*"*(2*i-1))
```

```python
# 20. Take input until user enters 0:
#     Count how many numbers are positive, negative, even, odd

pos = neg = even = odd = 0

while True:
    n = int(input("Enter number (0 to stop): "))

    if n == 0:
        break

    if n > 0:
        pos += 1
    else:
        neg += 1

    if n % 2 == 0:
        even += 1
    else:
        odd += 1

print("Positive:", pos)
print("Negative:", neg)
print("Even:", even)
print("Odd:", odd)
```

```python
# 21. Find second largest number in a list using loop (without sorting).

nums = [10,50,90,20,80]

first = second = -9999

for n in nums:
    if n > first:
        second = first
        first = n
    elif n > second and n != first:
        second = n

print("Second Largest:", second)
```

```python
# 22. Simulate ATM system:
#     Balance check
#     Deposit
#     Withdraw (check insufficient balance using if-else)

balance = 1000

while True:
    print("1 Balance")
    print("2 Deposit")
    print("3 Withdraw")
    print("4 Exit")

    ch = int(input("Choice: "))

    if ch == 4:
        break

    elif ch == 1:
        print("Balance =", balance)

    elif ch == 2:
        amt = int(input("Deposit amount: "))
        balance += amt

    elif ch == 3:
        amt = int(input("Withdraw amount: "))
        if amt <= balance:
            balance -= amt
        else:
            print("Insufficient Balance")
```

```python
# 23. Check if a string is palindrome ignoring spaces and case.

text = "Never Odd Or Even"

clean = text.replace(" ","").lower()

if clean == clean[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# 24. Create a number guessing game:
#     User keeps guessing until correct
#     Give hints: "Too High" / "Too Low"

secret = 7

while True:
    guess = int(input("Guess number: "))

    if guess > secret:
        print("Too High")
    elif guess < secret:
        print("Too Low")
    else:
        print("Correct Guess!")
        break
```

```python
# 25. Process a list of student marks:
#     Print grade:
#       ≥90 → A
#       ≥75 → B
#       ≥50 → C
#       else → Fail
#     Also count how many students passed/failed

marks = [95,80,60,30,77]

passed = 0
failed = 0

for m in marks:
    if m >= 90:
        grade = "A"
        passed += 1
    elif m >= 75:
        grade = "B"
        passed += 1
    elif m >= 50:
        grade = "C"
        passed += 1
    else:
        grade = "Fail"
        failed += 1

    print(m, "→", grade)

print("Passed:", passed)
print("Failed:", failed)
```

```python
# Combine everything:
# Build a program that:
#   * Takes multiple student records (name + marks)
#   * Assign grades using if-elif-else
#   * Store results in list
#   * Print topper using loop
#   * Allow user to exit using while

students = []

while True:
    name = input("Enter name (exit to stop): ")

    if name.lower() == "exit":
        break

    marks = int(input("Enter marks: "))

    if marks >= 90:
        grade = "A"
    elif marks >= 75:
        grade = "B"
    elif marks >= 50:
        grade = "C"
    else:
        grade = "Fail"

    students.append((name, marks, grade))

topper = students[0]

for s in students:
    if s[1] > topper[1]:
        topper = s

print("Topper:", topper)
```
